```python
import pandas as pd
from sklearn import preprocessing
import math
#-------------------------------------------------------------------------------------------------
#Import both dataframes, merge them and clean some features:
df = pd.read_csv("datasets/bigdf")
df1 = pd.read_csv("datasets/smalldf")


agedic = dict(zip(df1.ref, df1.age))
areadic = dict(zip(df1.ref, df1.area))
attrdic = dict(zip(df1.ref, df1.attr))
bathdic = dict(zip(df1.ref, df1.bath))
floordic = dict(zip(df1.ref, df1.floor))


df["age"] = df.ref.map(agedic)
df["area"] = df.ref.map(areadic)
df["attr"] = df.ref.map(attrdic)
df["bath"] = df.ref.map(bathdic)
df["floor"] = df.ref.map(floordic)

df["real state"] = df["real state"].replace("nan", "Prop")
df["real state"] = df["real state"].astype(str)

#Save everything in a csv file called mergedf:
df.to_csv("datasets/mergedf")

#------------------------------------------------------------------------------------
#Join the 2 dataframes and complete them:
df = pd.read_csv('datasets/mergedf')
df1 = pd.read_csv('datasets/coordapart')

dictlat = dict(zip(df1.ref, df1.lat))
dictlng = dict(zip(df1.ref, df1.lng))

df["lat"] = df.ref.map(dictlat)
df["lng"] = df.ref.map(dictlng)

df4 = df[["ref", "name", "rent", "gross_ar", "salar_ar", "real state", "bedroom", "bath", "age", "floor", "lat", "lng"]]
df4.to_csv("datasets/completdf")

#-------------------------------------------------------------------------------------------------
#Add another feature that with label the area code for each apartment, clean attribute column if we want to use it in the future
# using CountVectorizer from sklearn:

df = pd.read_csv("datasets/trabfin")

#Cleaning attribute feature:
limp = []
for i in df.attr:
    if i == "0":
        limp.append("Nope")
    else:
        x = i.replace("\r\r\r\n\r\r\r\n\r\r\r\n", "")
        t = x.replace("\r\r\r\n", " ")
        limp.append(t)
        
df["at_clean"] = limp

#Cleaning district feature:
district = []
for i in df.area:
    if i != 0:
        x = i.replace("\r\r\r\r\n", "")
        y = x.split(",")[-1].strip()
        district.append(y)
    else:
        district.append(i)
        
df["distrito"] = district

dictcoord= dict([(i,[a,b, c ]) for i, a,b, c in zip(df.ref, df.lat,df.lng, df.distrito)])

#Filling empty values for district => calculate the closest distance to another apartment and use his district data for that empty
#value:
lisdist = []
ref = []
for a,b, c in dictcoord.itervalues():
    if c == "0":
        lisdist1 = []
        for x,y,z in dictcoord.itervalues():
            if z != "0":
                dist = math.sqrt( (a - x)**2 + (b - y)**2 )
                if dist < 0.002:
                    lisdist1.append(z)
        lisdist.append(lisdist1[0])
        
   
    else:                 
        lisdist.append(c)
        
for i in dictcoord.iteritems():
    ref.append(i[0])
    
df["distrito"] = district

#Create a feature that for each district we have a label:
label = preprocessing.LabelEncoder()
etiqueta = label.fit_transform(df.distritofull)
df["district_label"] = etiqueta

#Save all the work - we will use this csv file to predict prices:
df.to_csv("datasets/workproject")
```
